```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
iris=load_iris()
```

```python
dir(iris)
```

```python
iris.feature_names
```

```python
df=pd.DataFrame(iris.data, columns=iris.feature_names)
```

```python
df.drop(['sepal length (cm)','sepal width (cm)'], axis='columns', inplace=True)
```

```python
df
```

```python
plt.scatter(df['petal length (cm)'],df['petal width (cm)'])
```

```python
km=KMeans(n_clusters=2)
km
```

```python
y_predicted=km.fit_predict(df[['petal length (cm)', 'petal width (cm)']])
y_predicted
```

```python
df['cluster']=y_predicted
df.head()
```

```python
df1=df[df.cluster==0]
df2=df[df.cluster==1]
```

```python
plt.scatter(df1['petal length (cm)'],df1['petal width (cm)'], color='green')
plt.scatter(df2['petal length (cm)'],df2['petal width (cm)'], color='red')
plt.xlabel('petal length (cm)')
plt.ylabel('petal width (cm)')
plt.legend()
```

```python
scaler=MinMaxScaler()
scaler.fit(df[['petal length (cm)']])
df['petal length (cm)']=scaler.transform(df[['petal length (cm)']])
df
```

```python
scaler.fit(df[['petal width (cm)']])
df['petal width (cm)']=scaler.transform(df[['petal width (cm)']])
df
```

```python
km=KMeans(n_clusters=3)
y_predicted=km.fit_predict(df[['petal length (cm)', 'petal width (cm)']])
y_predicted
```

```python
df['clus']=y_predicted
df
```

```python
km.cluster_centers_
```

```python
df1=df[df.clus==0]
df2=df[df.clus==1]
df3=df[df.clus==2]
```

```python
plt.scatter(df1['petal length (cm)'],df1['petal width (cm)'], color='green',label='sesitoa')
plt.scatter(df2['petal length (cm)'],df2['petal width (cm)'], color='red',label='versicolor')
plt.scatter(df3['petal length (cm)'],df3['petal width (cm)'], color='black',label='verginca')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color='yellow',marker='*',label='centeroid')
plt.xlabel('petal length (cm)')
plt.ylabel('petal width (cm)')
plt.legend()
```

```python
df
```

```python
k_rng=range(1,11)
sse=[]
for k in k_rng:
    km=KMeans(n_clusters=k)
    km.fit(df[['petal length (cm)','petal width (cm)']])
    sse.append(km.inertia_)
sse
```

```python
plt.xlabel('k')
plt.ylabel('sum of squared errors')
plt.plot(k_rng,sse)
```
